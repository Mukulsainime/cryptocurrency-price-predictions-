```python
def add_technical_indicators(df):
    #df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
    #df.set_index('timestamp', inplace=True)
    # RSI
    df['rsi'] = ta.momentum.RSIIndicator(df['close']).rsi()
    df['returns'] = df['close'].pct_change()
    df['price_range'] = df['high'] - df['low']
    #df['volume_ma'] = df['volume'].rolling(window=20).mean()
    # MACD
    macd = ta.trend.MACD(df['close'])
    df['macd'] = macd.macd()
    df['macd_signal'] = macd.macd_signal()
    #df['macd_diff'] = macd.macd_diff()

    # Bollinger Bands
    bollinger = ta.volatility.BollingerBands(df['close'])
    df['bollinger_hband'] = bollinger.bollinger_hband()
    #df['bollinger_lband'] = bollinger.bollinger_lband()
    #df['bollinger_mavg'] = bollinger.bollinger_mavg()

    # Volume indicators
    df['volume_adi'] = ta.volume.AccDistIndexIndicator(df['high'], df['low'], df['close'], df['volume']).acc_dist_index()
    #df['volume_obv'] = ta.volume.OnBalanceVolumeIndicator(df['close'], df['volume']).on_balance_volume()

    # Volatility
    df['volatility_atr'] = ta.volatility.AverageTrueRange(df['high'], df['low'], df['close']).average_true_range()
    #Get scaled differences between open and close
    df['delta'] = (df.close - df.open) / df.open
    # Adding a previous close column
    #df['prev_close'] = df['close'].shift(1)
    return df
```

```python
def add_time_features(df):
    df['hour'] = df.index.hour
    return df

#df = add_time_features(df)
```

```python
def add_roc_features(df):
    df['roc_short'] = ta.momentum.ROCIndicator(df['close'], window=12).roc()
    df['roc_medium'] = ta.momentum.ROCIndicator(df['close'], window=50).roc()
    df['roc_long'] = ta.momentum.ROCIndicator(df['close'], window=120).roc()
    return df
```

```python
def add_ema_features(df):
    df['ema_short'] = ta.trend.EMAIndicator(df['close'], window=10).ema_indicator()
    #df['ema_medium'] = ta.trend.EMAIndicator(df['close'], window=50).ema_indicator()
    #df['ema_long'] = ta.trend.EMAIndicator(df['close'], window=120).ema_indicator()
    return df

#df = add_ema_features(df)
```

```python
def add_lag_features(df, lags=[1,30,60]):
    for lag in lags:
        #df[f'close_lag_{lag}'] = df['close'].shift(lag)
        df[f'return_lag_{lag}'] = df['returns'].shift(lag)
    return df

#df = add_lag_features(df)
```

```python
import pandas as pd
import numpy as np

def calculate_iv(X, y):
    """
    Calculate the Information Value (IV) for each feature in X.
    
    Parameters:
    X : DataFrame of features
    y : Series of target variable
    
    Returns:
    iv_df : DataFrame with features and their corresponding IV values
    """
    
    iv_list = []  # List to store IV values
    
    # Loop through each feature
    for feature in X.columns:
        # Create a DataFrame with the feature and the target variable
        df = pd.concat([X[feature], y], axis=1)
        
        # Create bins for continuous variables (if necessary)
        # You can customize this step based on the type of feature
        df[feature] = pd.qcut(df[feature], q=10, duplicates='drop')  # You can adjust the number of bins
        
        # Calculate the count of Good and Bad (0 = Bad, 1 = Good)
        grouped = df.groupby(feature)[y.name].agg(['count', 'sum'])
        grouped['Bad'] = grouped['count'] - grouped['sum']
        
        # Calculate Distribution of Good and Bad
        grouped['Good Dist'] = grouped['sum'] / grouped['sum'].sum()
        grouped['Bad Dist'] = grouped['Bad'] / grouped['Bad'].sum()
        
        # Calculate WoE
        grouped['WoE'] = np.log(grouped['Good Dist'] / grouped['Bad Dist'])
        
        # Calculate IV for the feature
        grouped['IV'] = (grouped['Good Dist'] - grouped['Bad Dist']) * grouped['WoE']
        iv = grouped['IV'].sum()
        
        iv_list.append((feature, iv))
    
    # Convert the list of IV values into a DataFrame
    iv_df = pd.DataFrame(iv_list, columns=['Feature', 'IV'])
    iv_df = iv_df.sort_values(by='IV', ascending=False)  # Sort by IV values
    
    return iv_df

# Example usage
# Load your dataset (replace with your data)
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
X = train_df.drop('target', axis =1)
y = train_df.target

# Calculate IV for each feature
iv_df = calculate_iv(X, y)

# Print the IV values
print(iv_df)
```
